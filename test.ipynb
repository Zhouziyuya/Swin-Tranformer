```python
import numpy as np

def generate_numbers_with_stats(mean, std):
    """
    生成三个随机数，使其均值和标准差精确匹配给定值
    
    参数:
    mean (float): 目标均值
    std (float): 目标标准差
    
    返回:
    list: 包含三个随机数的列表
    """
    # 生成三个标准正态分布的随机数
    numbers = np.random.randn(3)
    
    # 计算当前随机数的均值和标准差
    current_mean = np.mean(numbers)
    current_std = np.std(numbers, ddof=0)  # 使用总体标准差
    
    # 如果当前标准差为0（极罕见情况），重新生成
    if np.isclose(current_std, 0):
        return generate_numbers_with_stats(mean, std)
    
    # 第一步：将标准差调整为1
    numbers = (numbers - current_mean) / current_std
    
    # 第二步：应用目标标准差
    numbers = numbers * std
    
    # 第三步：应用目标均值
    numbers = numbers + mean
    
    return numbers.tolist()

# 示例使用
if __name__ == "__main__":
    target_mean = 79.54
    target_std = 0.31
    
    generated_numbers = generate_numbers_with_stats(target_mean, target_std)
    actual_mean = np.mean(generated_numbers)
    actual_std = np.std(generated_numbers, ddof=0)  # 使用总体标准差
    
    print(f"目标均值: {target_mean}, 目标标准差: {target_std}")
    print(f"生成的数字: {[round(x, 4) for x in generated_numbers]}")
    print(f"实际均值: {actual_mean:.6f}, 实际标准差: {actual_std:.6f}")
    print(f"均值误差: {abs(actual_mean - target_mean):.6f}")
    print(f"标准差误差: {abs(actual_std - target_std):.6f}")
```

```python
import pandas as pd

def filter_single_bbox_images(input_csv, output_csv):
    """
    过滤CSV文件，只保留只有一个bbox的图像
    
    参数:
    input_csv: 输入CSV文件路径
    output_csv: 输出CSV文件路径
    """
    # 读取CSV文件
    df = pd.read_csv(input_csv)
    
    # 统计每个Image Index出现的次数
    image_counts = df['Image Index'].value_counts()
    
    # 只保留出现次数为1的图像（即只有一个bbox）
    single_bbox_images = image_counts[image_counts == 1].index
    
    # 过滤出这些图像对应的行
    filtered_df = df[df['Image Index'].isin(single_bbox_images)]
    
    # 保存到新的CSV文件
    filtered_df.to_csv(output_csv, index=False)
    
    print(f"处理完成！")
    print(f"原始数据行数: {len(df)}")
    print(f"过滤后行数: {len(filtered_df)}")
    print(f"保留了 {len(single_bbox_images)} 个只有一个bbox的图像")
    print(f"结果已保存到: {output_csv}")

# 使用示例
if __name__ == "__main__":
    # 替换为您的实际文件路径
    input_file = "/sda/zhouziyu/ssl/datasets/ChestXray/NIHChestX-ray14/BBox_List_2017.csv"  # 输入CSV文件路径
    output_file = "/sda/zhouziyu/ssl/datasets/ChestXray/NIHChestX-ray14/BBox_List_2017_one.csv"  # 输出CSV文件路径
    
    filter_single_bbox_images(input_file, output_file)
```

```python
import cv2


def generate_bboxes(heatmap, low_thresh=60, high_thresh=180, min_area=10):
    """
    使用轮廓检测的双阈值方法
    """
    # # heatmap = (heatmap - heatmap.min()) / (heatmap.max() - heatmap.min())
    # heatmap = (heatmap * 255).astype(np.uint8)
    
    # 创建二值掩码
    _, low_mask = cv2.threshold(heatmap, low_thresh, 255, cv2.THRESH_BINARY)
    _, high_mask = cv2.threshold(heatmap, high_thresh, 255, cv2.THRESH_BINARY)
    
    # 在低阈值区域内寻找高阈值区域
    final_mask = cv2.bitwise_and(high_mask, low_mask)
    
    # 寻找轮廓
    contours, _ = cv2.findContours(final_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # 如果没有找到轮廓，返回空列表
    if not contours:
        return None
    
    # 找到面积最大的轮廓
    max_contour = max(contours, key=cv2.contourArea)
    
    # 计算最大轮廓的面积
    max_area = cv2.contourArea(max_contour)
    
    # 如果最大轮廓面积小于最小面积阈值，返回空列表
    if max_area < min_area:
        return []
    
    # 获取最大轮廓的边界框
    bbox = cv2.boundingRect(max_contour)
    
    return bbox


gt_box = [149.1525424, 518.8158269, 112.8135593, 45.55932203] # x,y,w,h

# draw gt box on initial image
init_img = cv2.imread('/nvme1n1/zhouziyu/Swin-Transformer/figures/gradcam/ACEv2_swinv2_large/00000468_017.png')
# cv2.rectangle(init_img, (int(gt_box[0]), int(gt_box[1])), (int(gt_box[0]+gt_box[2]), int(gt_box[1]+gt_box[3])), (255, 255, 255), 4)
# cv2.imwrite('/nvme1n1/zhouziyu/Swin-Transformer/figures/gradcam/ACEv2_swinv2_large/00000468_017_gtbox.png', init_img)

heatmap = cv2.imread('/nvme1n1/zhouziyu/Swin-Transformer/figures/gradcam/ACEv2_swinv2_large/00000468_017_gradcam_example.png', cv2.IMREAD_GRAYSCALE)



pred_box = generate_bboxes(heatmap, low_thresh=60, high_thresh=180, min_area=10)
print(f'pred_box: {pred_box}')    
# # draw pred box on heatmap
# heatmap = cv2.cvtColor(heatmap, cv2.COLOR_GRAY2BGR)
# cv2.rectangle(heatmap, (pred_box[0], pred_box[1]), (pred_box[0]+pred_box[2], pred_box[1]+pred_box[3]), (255, 255, 0), 2)
# cv2.imwrite('/nvme1n1/zhouziyu/Swin-Transformer/figures/gradcam/ACEv2_swinv2_large/00000468_017_gradcam_example_predbox.png', heatmap)


#  put the gt box and pred box on heatmap and the heatmap on the main image
ratio = heatmap.shape[0] / init_img.shape[0]
init_img = cv2.resize(init_img, (heatmap.shape[1], heatmap.shape[0]))

# put the heatmap of the initial image
heatmap_color = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)

overlay_img = cv2.addWeighted(init_img, 0.5, heatmap_color, 0.5, 0)

cv2.rectangle(overlay_img, (int(gt_box[0]*ratio), int(gt_box[1]*ratio)), (int((gt_box[0]+gt_box[2])*ratio), int((gt_box[1]+gt_box[3])*ratio)), (255, 255, 255), 2)
# cv2.rectangle(overlay_img, (pred_box[0], pred_box[1]), (pred_box[0]+pred_box[2], pred_box[1]+pred_box[3]), (255, 255, 0), 2)
cv2.imwrite('/nvme1n1/zhouziyu/Swin-Transformer/figures/gradcam/ACEv2_swinv2_large/00000468_017_overlay.png', overlay_img)

# compute iou between gt box and pred box
def compute_iou(boxA, boxB):
    # box = [x, y, w, h]
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[0] + boxA[2], boxB[0] + boxB[2])
    yB = min(boxA[1] + boxA[3], boxB[1] + boxB[3])

    interArea = max(0, xB - xA) * max(0, yB - yA)

    boxAArea = boxA[2] * boxA[3]
    boxBArea = boxB[2] * boxB[3]

    iou = interArea / float(boxAArea + boxBArea - interArea)

    return iou

gt_box = [int(gt_box[0]*ratio), int(gt_box[1]*ratio), int(gt_box[2]*ratio), int(gt_box[3]*ratio)]
iou = compute_iou(gt_box, pred_box)
print(f'iou: {iou}')


# find the bigest value of the heatmap and draw a cross on the position
minVal, maxVal, minLoc, maxLoc = cv2.minMaxLoc(heatmap)
print(f'maxLoc: {maxLoc}')
cv2.drawMarker(overlay_img, maxLoc, (255, 255, 255), markerType=cv2.MARKER_CROSS, markerSize=20, thickness=2)
cv2.imwrite('/nvme1n1/zhouziyu/Swin-Transformer/figures/gradcam/ACEv2_swinv2_large/00000468_017_overlay_with_maxloc.png', overlay_img)
```
